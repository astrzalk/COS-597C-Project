```python
def getNotEmptyCols(df, thresh=0.75):
    good_cols = []
    for c in df:
        prct_empty = sum(df[c].isna())/len(df[c])
        if prct_empty < thresh:
            good_cols.append(c)
    return good_cols

pat_cols = []
for pat in docs:
    df = pd.DataFrame(pat['data_ts'])
    df.set_index('WCT',inplace=True)
    pat_cols.append(getNotEmptyCols(df))
```
